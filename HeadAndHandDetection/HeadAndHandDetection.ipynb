# Head and Hand detection based on skin colour from a given input image

```python
# import all the required libraries
import cv2
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# read an image using opencv and plotting it.
image = cv2.imread("image2.jpg")
plt.imshow(image,cmap="gray")
plt.show()
```

```python
# converting the rgb image into hsv color conversion

image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

#### Read more about HSV here: https://en.wikipedia.org/wiki/HSL_and_HSV

```python
# defining the threshold values to segment the skin region

lower_skin = np.array([0,40,0], dtype=np.uint8)
upper_skin = np.array([20,255,255], dtype=np.uint8)
```

```python
# Detect an object based on the range of pixel values in the HSV colorspace.

final_image = cv2.inRange(image_hsv, lower_skin, upper_skin)
```

```python
plt.imshow(final_image,cmap="gray")
plt.show()
```
